```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r'Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
df.plot()
```

```python
df.plot(kind='line')
```

```python
df.plot(kind='line', subplots=True)
```

```python
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Date')
```

```python
df.plot(kind='bar', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Date')
```

```python
df.plot(kind='bar', stacked=True, title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Date')
```

```python
df['Flavor Rating'].plot(kind='bar', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Date')
```

```python
df.plot.barh(stacked=True, xlabel='Daily Ratings')
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500)
```

```python
column_totals = df.sum()
column_totals 
```

```python
column_totals.plot.pie(labels=column_totals, title='Distribution of ratings',figsize=(7,7))

plt.legend(title='Ratings', loc='best')
```
